```python
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

```python
x=np.array([2023, 2024, 2025, 2026])
y=np.array([25,30,20,35])
plt.plot(x,y, marker=".",ms=30,markerfacecolor="cyan",markeredgecolor="#1cd3fc",
         linestyle="solid",
         linewidth="3",color="#000080")
plt.show()
```

```python
x=np.array([2023, 2024, 2025, 2026])
y1=np.array([25,30,20,35])
y2=np.array([17,23,38,51])
y3=np.array([12,21,20,18])
line_style= dict( marker=".",ms=30,markerfacecolor="cyan",markeredgecolor="#1cd3fc",
         linestyle="solid",
         linewidth="3",color="#000080")

plt.plot(x,y1,**line_style)
plt.plot(x,y2,**line_style)
plt.plot(x,y3,**line_style)
plt.show()
```

```python
x=np.array([2023, 2024, 2025, 2026])
y1=np.array([25,30,20,35])
y2=np.array([17,23,38,51])
y3=np.array([12,21,20,18])
line_style= dict( marker=".",ms=30,markerfacecolor="cyan",markeredgecolor="#1cd3fc",
         linestyle="solid",
         linewidth="3")

plt.plot(x,y1,color="#008000", **line_style)
plt.plot(x,y2,color="#FFFF00", **line_style)
plt.plot(x,y3,color="#FF0000", **line_style)
plt.show()
```

```python
from matplotlib import color_sequences
x=np.array([2023, 2024, 2025, 2026])
y1=np.array([25,30,20,35])
y2=np.array([17,23,38,51])
y3=np.array([12,21,20,18])
plt.title("class size",fontsize=20,family='Arial',fontweight="bold",color="#2d4cfc")
plt.xlabel("year",fontsize=20,family="Arial",fontweight="bold",color="#2d4cfc")
plt.ylabel("students ",fontsize=20,family="Arial",fontweight="bold",color="#2d4cfc")

plt.tick_params(axis="both",labelsize=15,color="#2d4cfc")


plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xticks(x)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#bar charts
categories=np.array(["grains","fruits","vegetable","protien","dairy","sweets"])
values=np.array([5,3,6,7,2,8])
plt.bar(categories,values,color="#000080")
#for horizontal bar chart
#plt.barh(categories,values,color="#000080")
plt.title("daily consumption")
plt.xlabel("food")
plt.ylabel("quantity")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
categories=np.array(["freshman","sophomore","junior","senior"])
values=np.array([300,350,250,290])
colors=["red","blue","green","yellow"]
plt.pie(values,labels=categories,colors=colors,autopct="%1.1f%%",
        explode=[0,0,0,0.2],
        shadow=True,
        startangle=90)
plt.title("hardik's college")
```

```python
#scatter graphs
import numpy as np
import matplotlib.pyplot as plt
x1=np.array([0,1,1,2,3,4,5,5,6,8,10]) #hours studies
y1=np.array([55,60,65,62,70,66,87,82,89,78,72]) #marks
x2=np.array([0,1,2,2,3,4,5,6,7,8,8])
y2=np.array([50,58,60,76,56,84,94,78,90,77,88])


plt.xlabel("hours studies")
plt.ylabel("marks")
plt.title("hours vs marks")
plt.scatter(x1,y1,color="skyblue",alpha=.5,s=200,label="class A")
plt.scatter(x2,y2,color="red",alpha=.5,s=200,label="class B")
plt.legend()
plt.show()
```

```python
 #histogram
import numpy as np
import matplotlib.pyplot as plt
scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)
plt.hist(scores,bins=10,color="lightgreen",edgecolor="black")
plt.title("test scores")
plt.xlabel("score")
plt.ylabel("no. of students")
plt.show()
```

```python
#subplts
import numpy as np
import matplotlib.pyplot as plt
figure,axes=plt.subplots(2,2)
x=np.array([1,2,3,4,5])
axes[0,0].plot(x,x**2,color="red")
axes[0,0].set_title("square")
axes[0,1].plot(x,x**3,color="green")
axes[0,1].set_title("cube")
axes[1,0].plot(x,x**4,color="blue")
axes[1,0].set_title("power")
axes[1,1].plot(x,x**5,color="yellow")
axes[1,1].set_title("power")
plt.tight_layout()
plt.show()
```

```python
#pandas + matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("pokemon.csv")
type_count=(df["Type1"].value_counts(ascending=True))
plt.barh(type_count.index, type_count.values,color="yellow",edgecolor="black")
plt.xlabel("count")
plt.ylabel("type")
plt.title("pokemon type")
plt.tight_layout()
plt.show()
```
