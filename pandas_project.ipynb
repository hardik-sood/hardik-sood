```python
# ============================================================
#  Drug Classification Dataset - Exploratory Data Analysis
# ============================================================

# Author: Hardik Sood
# Description: Exploratory Data Analysis (EDA) of a pharmaceutical dataset
# Libraries used: pandas, matplotlib, seaborn

# -----------------------------
# STEP 1: Import Libraries
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting
%matplotlib inline

# -----------------------------
# STEP 2: Upload and Load Dataset
# -----------------------------
from google.colab import files
uploaded = files.upload()

# After upload, read the CSV file (replace filename if needed)
df = pd.read_csv("drug200.csv")

# -----------------------------
# STEP 3: Basic Exploration
# -----------------------------
print("Dataset Shape:", df.shape)
print("\nDataset Info:\n")
df.info()

print("\nFirst 5 Rows:")
display(df.head())

# -----------------------------
# STEP 4: Statistical Summary
# -----------------------------
print("\nSummary Statistics:")
display(df.describe(include='all'))

# -----------------------------
# STEP 5: Check Missing Values
# -----------------------------
print("\nMissing Values per Column:")
print(df.isnull().sum())

# -----------------------------
# STEP 6: Analyze Categorical Columns
# -----------------------------
print("\nUnique Values in Each Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

# -----------------------------
# STEP 7: Data Visualization
# -----------------------------

# Distribution of target variable (Drug Type)
plt.figure(figsize=(6,4))
sns.countplot(x='Drug', data=df, palette='viridis')
plt.title("Distribution of Drug Types")
plt.show()

# Gender vs Drug Type
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Drug', data=df, palette='mako')
plt.title("Drug Type by Gender")
plt.show()

# Age vs Blood Pressure vs Drug
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Na_to_K', hue='Drug', data=df, palette='tab10')
plt.title("Age vs Na_to_K by Drug Type")
plt.show()

# -----------------------------
# STEP 8: Correlation Analysis
# -----------------------------
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# STEP 9: Insights Summary
# -----------------------------
print(" Key Insights:")
print("""
1. The dataset has {} entries and {} columns.
2. Drug distribution shows which drugs are most prescribed.
3. Gender and blood pressure level may influence drug type.
4. Na_to_K ratio appears important for drug classification.
""".format(df.shape[0], df.shape[1]))

# -----------------------------
# STEP 10: Save Cleaned Dataset (Optional)
# -----------------------------
df.to_csv("cleaned_drug200.csv", index=False)
print(" Cleaned dataset saved as 'cleaned_drug200.csv'")
```
